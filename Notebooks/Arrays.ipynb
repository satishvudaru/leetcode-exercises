```python
#TwoSum using bruteforce

nums =[2,7,11,15]
target =18

for i in range(len(nums)):
    for j in range(1,len(nums)):
        if nums[i]+nums[j] ==target:
            print (f'i:{i}, j:{j}')
            break;
```

```python
# Group anagrams
# practice this again contans mulitple syntaxes
from collections import defaultdict

def group_anagrams(lstWords):

    final_dict={}
    for word in lstWords:
        sorted_letters = sorted(word)
        sorted_word = ''.join(sorted_letters)
        if sorted_word in final_dict:
            final_dict[sorted_word].append(word)
        else:
            final_dict[sorted_word] =[word]

    return final_dict.values()

words=["eat","tea","tan","ate","nat","bat"]
group_anagrams(words)
```

```python
#Contains duplicates
def contains_duplicates(nums):
    seen_set=set()
    for num in nums:
        if num in seen_set:
            return True
        
    return False   

nums =[1,2,3,4,5,6]
contains_duplicates(nums)
#space complexity is O(n) since set stores n words, the input can be n no of elements in the list.
```

```python
#Find K frequent items

from collections import Counter

def topKFrequent(nums, k):
    count = Counter(nums)

    return [
        num for num, freq in
        sorted(count.items(), key=lambda x: x[1], reverse=True)[:k]
    ]
```
